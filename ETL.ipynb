```python
import pandas as pd
```

## Step 1: Extract - Load the data from the CSV file


```python
# Load the dataset
file_path = "train.csv"  
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print("Extracted Data:")
data.head()
```

```python
# Drop columns that are not needed
columns_to_drop = ["PassengerId", "Name", "Ticket", "Cabin"]
data_transformed = data.drop(columns=columns_to_drop)
data_transformed
```

```python
# Fill missing values in the 'Age' column with the median age
data_transformed["Age"].fillna(data_transformed["Age"].median(), inplace=True)
data_transformed
```

```python
# Create a new column 'FamilySize' by combining 'SibSp' and 'Parch'
data_transformed["FamilySize"] = data_transformed["SibSp"] + data_transformed["Parch"]
data_transformed
```

## Step 3: Load - Save the transformed data to a new CSV file or a SQL database or Cloud Data storage


```python
#Save to a new CSV file
output_csv_path = "transformed_data.csv"
data_transformed.to_csv(output_csv_path, index=False)
print(f"\nTransformed data saved to {output_csv_path}")
```
